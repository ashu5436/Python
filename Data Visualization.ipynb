```python
ds=sns.load_dataset('attention')
ds
```

```python
plt.title('Scatter Plot')
sns.scatterplot(ds['score'],ds['solutions'])
plt.show()
```

```python
plt.title('rel Plot')
sns.relplot(ds['score'],ds['solutions'])
plt.show()
```

```python
plt.title('line Plot')
sns.lineplot(ds['score'],ds['solutions'])
plt.show()
```

```python
plt.title('dist Plot')
sns.distplot(ds['score'])
plt.show()
```

```python
plt.title('kde Plot')
sns.kdeplot(ds['score'],ds['solutions'])
plt.show()
```

```python
plt.title('hist Plot')
sns.histplot(ds['score'],bins=10)
plt.show()
```

```python
plt.title('joint Plot')
sns.jointplot(ds['score'],ds['solutions'])
plt.show()
```

```python
plt.title('violin Plot')
sns.violinplot(ds['score'])
plt.show()
```

```python
plt.title('rug Plot')
sns.rugplot(ds['solutions'],ds['score'])
plt.show()
```

```python
plt.title('pair Plot')
sns.pairplot(ds)
plt.show()
```

```python
plt.title('Strip Plot')
sns.stripplot(ds['score'],ds['solutions'])
plt.show()
```

```python
plt.title('Swarm Plot')
sns.swarmplot(ds['score'],ds['solutions'])
plt.show()
```

```python
plt.title('count Plot')
sns.countplot(ds['attention'])
plt.show()
```

```python

sns.catplot('score','solutions',hue='attention',data=ds,kind='bar')
```

```python
plt.title('box Plot')
sns.boxplot(ds['score'])
plt.show()
```

```python
plt.title('lm Plot')
sns.lmplot('score','solutions',hue='attention',data=ds)
plt.show()
```
